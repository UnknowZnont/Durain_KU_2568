```python
# get all filenames (without extension) under 'train'
from pathlib import Path


file_names_train = {p.stem: p for p in Path('train/images').rglob('*') if p.is_file()}
file_names_test  = {p.stem: p for p in Path('test/images').rglob('*')  if p.is_file()}
file_names_val   = {p.stem: p for p in Path('valid/images').rglob('*') if p.is_file()}

# merge into one dict mapping filename (stem) -> Path (path as value)
file_names = {}
file_names.update(file_names_val)
file_names.update(file_names_test)
file_names.update(file_names_train)

print(f"Total images: {len(file_names)}")
```

```python
labelData = {p.stem: p.read_text(encoding='utf-8', errors='ignore').splitlines() for p in labels}
print(f"Read {len(labelData)} label files")
```

```python
labelDataNotEmpty = {k: v for k, v in labelData.items() if len(v) > 0}
print(f"Label files with content: {len(labelDataNotEmpty)}")
```

```python
labelDataMoreThanOne = {k: v for k, v in labelData.items() if len(v) > 1}
print(f"Label files with more than one object: {len(labelDataMoreThanOne)}")
```

```python
labelDataOnlyOne = {k: v for k, v in labelData.items() if len(v) == 1}
print(f"Label files with only one object: {len(labelDataOnlyOne)}")
```

```python
labelDataOnlyOneByClass = {}
for k, v in labelDataOnlyOne.items():
    class_id = v[0].split()[0]  # get the class id from the single line
    if class_id not in labelDataOnlyOneByClass:
        labelDataOnlyOneByClass[class_id] = []  # initialize list if class_id not present
    labelDataOnlyOneByClass[class_id].append(k)  # add the file name to the class_id list

labelDataOnlyOneByClass
```

```python
def copy_image_and_label(label_stems, labels_list, file_names_map, dest_root='selected/more_than_one'):
    import shutil
    dest_root = Path(dest_root)
    images_out = dest_root / 'images'
    labels_out = dest_root / 'labels'
    images_out.mkdir(parents=True, exist_ok=True)
    labels_out.mkdir(parents=True, exist_ok=True)

    label_paths = {p.stem: p for p in labels_list}

    copied_images = 0
    copied_labels = 0
    for stem in label_stems:
        img_path = file_names_map.get(stem)
        if img_path and img_path.is_file():
            shutil.copy2(img_path, images_out / img_path.name)
            copied_images += 1

        lab_path = label_paths.get(stem)
        if lab_path and lab_path.is_file():
            shutil.copy2(lab_path, labels_out / lab_path.name)
            copied_labels += 1

    print(f"Copied {copied_images} images and {copied_labels} labels to '{dest_root.resolve()}'")
    return copied_images, copied_labels
```

```python
copy_image_and_label(labelDataMoreThanOne.keys(), labels, file_names, 'selected/more_than_one')
```

```python
copy_image_and_label(labelDataOnlyOne.keys(), labels, file_names, 'selected/only_one')
```

```python
for k, v in list(labelDataOnlyOneByClass.items()):
   copy_image_and_label(v, labels, file_names, 'selected/only_one_by_class_' + k)
```

```python
def split_file_into_train_test_val(file_list, train_ratio=0.7, test_ratio=0.2, val_ratio=0.1, seed=42):
    import random
    random.seed(seed)
    random.shuffle(file_list)

    total = len(file_list)
    train_end = int(total * train_ratio)
    test_end = train_end + int(total * test_ratio)

    train_files = file_list[:train_end]
    test_files = file_list[train_end:test_end]
    val_files = file_list[test_end:]

    return train_files, test_files, val_files
```

```python
split_file_into_train_test_val(list(labelDataOnlyOne.keys()), 0.7, 0.2, 0.1)
train_files, test_files, val_files = split_file_into_train_test_val(list(labelDataOnlyOne.keys()), 0.7, 0.2, 0.1)

print(f"Train: {len(train_files)}, Test: {len(test_files)}, Val: {len(val_files)}")

copy_image_and_label(train_files, labels, file_names, 'selected/split_only_one/train')
copy_image_and_label(test_files, labels, file_names, 'selected/split_only_one/test')
copy_image_and_label(val_files, labels, file_names, 'selected/split_only_one/val')
```
